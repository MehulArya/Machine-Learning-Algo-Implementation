Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
from sklearn.datasets import make_regression
```

Create Synthetic Regressor Data

```python
x,y = make_regression(n_samples=100, n_features=1, noise=15, random_state=42)
print("x: \n",x)
print("y: \n",y)
```

Initialize the SGD Regressor (LMS).
SGDRegressor uses stochastic gradient descent to learn weights

```python
sgd_reg = SGDRegressor(max_iter=1000, eta0=0.1, learning_rate='invscaling', random_state=42)
```

Train the Model

```python
sgd_reg.fit(x,y)
```

Plot the data and regression line

```python
plt.scatter(x, y, color='blue', label='Training Data')
y_pred = sgd_reg.predict(x) # Predicted Regressor Line
plt.plot(x, y_pred, color='green', label='SGD Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
